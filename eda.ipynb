```python
# 도메인 지식으로 frontend라고 구분지어도되나
for user_grp in user_tags_merge_grp:
    print(user_grp)
    # 관심사 키워드 유니크값만
    interest_keyword = ' '.join(user_grp[1]['keyword'].unique())
    user_tags.loc[user_tags.userID ==user_grp[0],'interest_keyword'] = interest_keyword
```
